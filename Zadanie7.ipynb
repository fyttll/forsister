```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\User\Desktop\test\forsister\web_clients_correct.csv"

df = pd.read_csv(file_path)
print(df.head())

plt.scatter(df['age'], df['bill'])
plt.xlabel('Age')
plt.ylabel('Bill')
plt.title('Scatterplot of Age vs Bill (matplotlib)')
plt.show()

sns.scatterplot(data=df, x='age', y='bill')
plt.title('Scatterplot of Age vs Bill (seaborn)')
plt.show()

sns.jointplot(x='age', y='bill', data=df, kind='reg')
plt.show()
```
